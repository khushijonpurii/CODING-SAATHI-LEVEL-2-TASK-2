```python
#/content/Iris.csv
```

```python
# create dataframe
import pandas as pd
df=pd.read_csv('/content/Iris.csv')
df
```

```python
#divide the data into ip and op
```

```python
x=df.iloc[:,1:5].values   #input except species
x
```

```python
y=df.iloc[:,5].values     #output species
y
```

```python
#checking for unique class names
import numpy as np
n=np.unique(y)
n
```

```python
#train and test variables
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
print(x.shape)
print(x_train.shape)  #75%
print(x_test.shape)   #25%
```

```python
print(y.shape)
print(y_train.shape)  #75%
print(y_test.shape)   #25%
```

```python
#normalisation or scaling of the iris datset
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train= scaler.fit_transform(x_train)
x_test= scaler.fit_transform(x_test)
```

```python
#applying decision tree classifier
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
#fit the model
model.fit(x,y)
```

```python
#visualisation
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
fig,ax=plt.subplots(figsize=(12, 12))
tree.plot_tree(model,feature_names=df.columns[1:5],class_names=n,filled=True,ax=ax)
plt.show()
```

```python
#predict the output
```

```python
y_pred=model.predict(x_test)
y_pred                           #predicted value
```

```python
y_test            #actual value
```

```python
#accuracy
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
